*For a Watts-Strogatz network with 100 nodes iterate over the range of the rewiring probabilty $p \in [0,1],$ using reasonable many steps in between, and compute the average shortest path length for all rewiring probabilties within this range (degree parameters $k=4$ and $k=8$). Plot both results in one figure with a legend and give both curves a customized color. Estimate the asymptotic spl-value for $p \rightarrow 1$ and plot them as dashed lines.*

```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
plt.style.use("seaborn-notebook")
```

```python
n = 100
ks = [4, 8]
ps = np.linspace(0, 1, 1001)

for k in ks:
    asympt_spl = np.log(n) / np.log(k)
    plt.plot(ps, asympt_spl*np.ones(len(ps)), color="grey", linestyle="dashed")

    spls = [nx.average_shortest_path_length(nx.watts_strogatz_graph(n, k, p)) for p in ps]
    plt.plot(ps, spls, label=f"k = {k}")

plt.title("average shortest path length in Watts-Strogatz model")
plt.xlabel("rewiring probability")
plt.legend()
plt.show()
```
